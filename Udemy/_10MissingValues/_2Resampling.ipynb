```python
import numpy as np
import pandas as pd
```

```python
# Sallayarak bir kanser dataası oluşturalım. 900 tane 0, 100 tane 1 olan dataset

np.random.seed(42)

set1no = 900
set2no = 100
```

```python
df1 = pd.DataFrame({
    "feature_1": np.random.normal(loc=0, scale=1, size=set1no),
    "feature_2": np.random.normal(loc=0, scale=1, size=set1no),
    "Target": 0
    
})

df2 = pd.DataFrame({
    "feature_1": np.random.normal(loc=0, scale=1, size=set2no),
    "feature_2": np.random.normal(loc=0, scale=1, size=set2no),
    "Target": 1
})

df1
```

```python
df = pd.concat([df1, df2]).reset_index(drop=True)
df
```

# **Upsampling**
* Eğer az datmız varsa bunu kullanarak artırabiliriz ama negatif yanları var çünkü aslında yeni dataları uyduruyorsun.
* Tabiki kafadan uydurmuyorsun belli bir algoritmaya göre çoğaltıyorsun, ama yine de gerçek data değil. Abartmamak lazım
# **Downsampling**
* Çok datamız varsa, data sayımızı azaltabiliriz. Her zaman tercih edilmeyebilir

#### Önce **Upsampling** ile başlayalım:

```python
# Upsampling:
df_minority = df[df["Target"] == 1]
df_minority.head()

df_majority = df[df["Target"] == 0]
```

```python
from sklearn.utils import resample
```

```python
df_minority_upsampled = resample(df_minority, replace=True, n_samples=900, random_state=42)
df_minority_upsampled
# Bunu çalıştırınca biz kanser olan elimizdeki verilerin sayısını 100'den 900'e çıkartmış olacağız.
```

```python
df_upsampled = pd.concat([df_majority, df_minority_upsampled])
df_upsampled
```

### Downsampling

```python
df_majority_downsampled = resample(df_majority, replace=True, n_samples=100, random_state=42)
df_majority_downsampled
```
