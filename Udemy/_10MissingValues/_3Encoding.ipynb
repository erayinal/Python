# **Encoding**
* Metin halindeki değerleri numarasallaştırmaya denir.
* Çünkü bilgisayar için anlamlı bir hale gelmesi lazım verinin
* Bunların çeşitleri var:
  * One-Hot Encoding
  * Label Encoding
  * Ordinal Encoding
  * Frequency Encoding
  * Target Encoding

## **One-Hot Encoding**
* 
* Ama eğer çok fazla feature varsa verinin bunu yapmak mantıklı değildir.

## Label Encoding
* Burada ise her değere bir sayı atarız
* 
* bunları mantıklı yerde kullanmak lazım, yani aslında derecelendirme yapılabilecek yerlerde kullanmak daha mantıklı.
* Mesela bu örnek mantıklı çünkü sayı ile birlikte boyut büyüyor ama bunları renkler için yapmak mantıklı değil çünkü nilgisayar bunu anlamaya çalışacak arasında ne gibi bir ilişki var diye.

## Ordinal Encoding
* Burada da yukarıdaki gibi derecelendirme olur, mantıklı olan budur
* Sıralamayı kendin yaparsın
* 

## Frequency Encoding
* Burada en çok görülme sıklığına göre değiştirebiliriz verdiğimiz değeri.
* Mesela bir terimden ne kadar çok geçiyorsa onun değeri daha yüksek olur ve bu aslında ikisi arasında bir anlam da barındırır
*

* Ortalama ile numaralandırma yaparız.
