## **Dış Kaynaklı Veri Okumak**

#### **.txt ve .csv Formatı:**
* Burada SQL ortamları dışında en çok karşılaşacağımız **'.txt'**, **'.csv'**, **'.xlsx'** 3 veri formatını da ele alacağız
* Bu işlem için 'Pandas' içerisinden bazı fonksiyonlar gerekli
* **.txt** ve **.csv** formunu aynı fonkisiyon ile okuyabiliyoruz:
  > txt ve cvs formatındaki veri setleri genelde ';' ile ayrılır ama bizim default ayracımız ','dür. Yani iki tane parametre vereceğiz:
  >> İlk parametre dosyanın uzantısı(pwd), ikinci parametre ise 'sep' parametresidir:
  >>> pd.read_csv("reading_data/ornekcsv.csv", sep = ';') <br>
      pd.read_csv("reading_data/duz_metin.txt")
* Yukarıda txt dosyasının orijanilinde ' ' boşluk işe ayrılmıştır ona rağmen ayrımı düzgün yapabilmiştir. Çünkü boşluğu da okuyabilmektedir her ne kadar default değeri ',' de olsa


#### **.xlsx Formatı(Excel):**
* Burada ise yine pandas'tan **'read_excel'** fonksiyonunu kullanacağız
  > pd.read_excel("reading_data/ornekx.xlsx")
